# TutorNow Data Cleaning & Feature Engineering

Load dataset and inspect structure

```
import pandas as pd

df = pd.read_csv('tutornow_dataset.csv')
df.head()
```

Convert date columns

```
df['session_date'] = pd.to_datetime(df['session_date'])
df['signup_date'] = pd.to_datetime(df['signup_date'])
```

Handle missing values (group-based imputation by tutor)

```
df['student_rating_given'] = df['student_rating_given'].fillna(
    df.groupby('tutor_id')['student_rating_given'].transform('mean')
)

df['student_rating_given'] = df['student_rating_given'].fillna(
    df['student_rating_given'].mean()
)
```

Feature Engineering

```
df['revenue'] = df['session_price']
df['is_completed'] = (df['session_status'] == 'Completed').astype(int)

df['year'] = df['session_date'].dt.year
df['month'] = df['session_date'].dt.month
df['day_of_week'] = df['session_date'].dt.day_name()

df['price_per_hour'] = df['session_price'] / df['session_duration_hours']
df['repeat_flag'] = df['repeat_student'].map({'Yes':1,'No':0})
```

Remove invalid records

```
df = df[df['session_price'] > 0]
```

Export cleaned dataset

```
df.to_csv('tutornow_cleaned.csv', index=False)
```
